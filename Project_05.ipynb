```python
#Relation Between Temperature, Humidity and Pressure

import matplotlib.pyplot as plt

import pandas as pd

import seaborn as sns

data = pd.read_csv('environmental_data.csv')
df = pd.DataFrame(data)

# Plotting pairwise relationships using seaborn
sns.pairplot(df[["Temperature", "Humidity", "Pressure"]],diag_kind='kde')

# Adding a title to the plot
plt.suptitle("Pairwise Relationships Among Temperature, Humidity, and Pressure", y=1.123)

plt.show()
```

```python
#Line Graph and Bar Graph

import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("environmental_data.csv")

# Display the first few rows of the data
print(df.head())

# Plot 1: Line Graph for Temperature, Humidity, and Pressure
plt.figure(figsize=(10, 6))

plt.plot(df.index, df["Temperature"], label="Temperature", marker='o', color="blue")

plt.plot(df.index, df["Humidity"], label="Humidity", marker='s', color="green")

plt.plot(df.index, df["Pressure"], label="Pressure", marker='^', color="red")

plt.title("Line Graph: Temperature, Humidity, and Pressure Over Time")

plt.xlabel("Index")

plt.ylabel("Values")

plt.legend()

plt.grid(alpha=0.5)

plt.show()

#  Bar Graph for Temperature, Humidity, and Pressure
plt.figure(figsize=(10, 6))
bar_width = 0.25
x = df.index

# Create side-by-side bars
plt.bar(x - bar_width, df["Temperature"], width=bar_width, label="Temperature", color="blue")

plt.bar(x, df["Humidity"], width=bar_width, label="Humidity", color="green")

plt.bar(x + bar_width, df["Pressure"], width=bar_width, label="Pressure", color="red")

plt.title("Bar Graph: Temperature, Humidity, and Pressure")

plt.xlabel("Index")

plt.ylabel("Values")

plt.legend()

plt.grid(alpha=0.3)

plt.show()
```

```python
#Line graph using bokeh

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


df=pd.read_csv('/content/environmental_data.csv')

from bokeh.plotting import figure,show ,output_notebook
from bokeh.models import HoverTool
import pandas as pd

output_notebook()

p=figure(title='simple line example',x_axis_label='Temperature',y_axis_label='Humidity')

p.line(df['Temperature'],df['Humidity']) # Changed df(['Humidity']) to df['Humidity']

p.scatter(df['Temperature'],df['Humidity'],size=8,color='red') # Changed [Temperature] to df['Temperature'] and ['Humidity'] to df['Humidity']

hover=HoverTool()
hover.tooltips=[('x','$x'),('y','$y')]
p.add_tools(hover)

show(p)
```

```python
#Pie_chart

import matplotlib.pyplot as plt

# Data
Temperature = [23, 39, 8, 46, 10, 39, 29, 46, -2, 35, 34, 14, 21, 21, 14, 37, 45, 18, 12, 13]
Humidity = [63, 87, 8, 50, 100, 97, 21, 18, 28, 44, 80, 29, 67, 8, 14, 77, 41, 60, 42, 92]
Pressure = [947, 1073, 941, 930, 1083, 991, 1074, 1018, 906, 982, 1068, 1005, 1007, 906, 910, 958, 945, 995, 1064, 1027]

# Calculate the sum of each list
sum_x = sum(Temperature)
sum_y = sum(Humidity)
sum_z = sum(Pressure)

# Pie chart data
labels = ['Temperature', 'Humidity', 'Pressure']
sizes = [sum_x, sum_y, sum_z]
colors = ['#ff9999','#66b3ff','#99ff99']  # Colors for the slices

# Create the pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=45)

# Equal aspect ratio ensures that pie is drawn as a circle.
plt.axis('equal')

# Add a title
plt.title('Pie Chart of Sum of x, y, and z')

# Show the plot
plt.show()
```

```python
# using seaborn on bar graph

import pandas as pd
import seaborn as  sns
import matplotlib.pyplot as plt
data=pd.read_csv('/content/environmental_data.csv')

df=pd.DataFrame(data)
sns.set_style('darkgrid')
sns.set_palette('muted')
sns.barplot(x='Temperature',y='Humidity',data=df)

plt.show()
```
